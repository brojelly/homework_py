__[과제 3: 주문 처리 메서드 process_order]__

목표: process_order 함수를 구현하여 손님의 주문을 처리하고 재고를 업데이트합니다.

설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
class BungeoppangShop:
    def __init__(self):
        #초기 재고 설정
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5}
        
        #가격 설정
        self.prices = {
            "팥붕어빵":1000,
            "슈크림붕어빵":1200,
            "초코붕어빵":1500}
        
        #판매량
        self.sales = {
            "팥붕어빵":0,
            "슈크림붕어빵":0,
            "초코붕어빵":0}

    def check_stock(self):
        print("현 재고:")
        for 빵, 재고 in self.stock.items():
            print(f'- {빵}: {재고}개 남음')

    def process_order(self, 빵_타입, 빵_숫자):
        #존재 X 붕어빵 주문 처리
        if 빵_타입 not in self.stock:
            print(f"{빵_타입}? 그런거 안팜")
            return
        
        #재고 부족
        if self.stock[빵_타입] < 빵_숫자:
            print(f'{빵_타입} 그건 다나갔어요(남은수량: {self.stock[빵_타입]}개)')
            return

        #주문이 가능한 경우: 재고 차감, 판매량 증가
        self.stock[빵_타입] -= 빵_숫자
        self.sales[빵_타입] += 빵_숫자
        total_price = self.prices[빵_타입] * 빵_숫자
        print(f" {빵_타입} {빵_숫자}개 주문 완료! 총 {total_price}원입니다.")
        print(f"남은 재고: {빵_타입} {self.stock[빵_타입]}개")
# 붕어빵 가게 객체 생성
shop = BungeoppangShop()

# 현재 재고 확인
shop.check_stock()

print("\n=== 주문 테스트 ===")
shop.process_order("팥붕어빵", 3)  # 정상 주문
shop.process_order("슈크림붕어빵", 10)  # 재고 부족
shop.process_order("고구마붕어빵", 2)  # 없는 메뉴 주문

# 주문 후 재고 확인
print("\n=== 주문 후 재고 확인 ===")
shop.check_stock()


```

- 프로그램이 종료될 때 총 매출을 계산하고 데이터를 저장합니다.


```python
import json  # 매출 데이터를 저장하기 위한 모듈

class BungeoppangShop:
    def __init__(self):
        # 초기 재고 설정
        self.stock = {
            "팥붕어빵": 10,
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }

        # 가격 설정
        self.prices = {
            "팥붕어빵": 1000,
            "슈크림붕어빵": 1200,
            "초코붕어빵": 1500
        }

        # 판매량 초기화
        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0
        }

    # 재고 확인 메서드
    def check_stock(self):
        print("\n📢 현재 붕어빵 재고:")
        for 빵, 재고 in self.stock.items():
            print(f'- {빵}: {재고}개 남음')

    # 주문 처리 메서드
    def process_order(self, 빵_타입, 빵_숫자):
        if 빵_타입 not in self.stock:
            print(f"❌ {빵_타입}? 판매하지 않는 품목입니다!")
            return
        
        if self.stock[빵_타입] < 빵_숫자:
            print(f"⚠️ {빵_타입} 재고가 부족합니다! (남은 수량: {self.stock[빵_타입]}개)")
            return

        self.stock[빵_타입] -= 빵_숫자
        self.sales[빵_타입] += 빵_숫자
        total_price = self.prices[빵_타입] * 빵_숫자

        print(f"✅ {빵_타입} {빵_숫자}개 주문 완료! 총 {total_price}원입니다.")
        print(f"🛒 남은 재고: {빵_타입} {self.stock[빵_타입]}개")

    # 관리자 모드 (재고 추가)
    def admin_mode(self, 빵_타입, 추가_재고):
        if 빵_타입 not in self.stock:
            print(f"❌ {빵_타입}은(는) 판매하는 품목이 아닙니다.")
            return

        if 추가_재고 <= 0:
            print("⚠️ 추가할 재고 수량은 1개 이상이어야 합니다!")
            return

        self.stock[빵_타입] += 추가_재고
        print(f"🔧 {빵_타입}의 재고가 {추가_재고}개 추가되었습니다. (총 {self.stock[빵_타입]}개)")

    # ✅ 매출 계산 메서드
    def calculate_total_sales(self):
        total_sales = 0  # 총 매출 변수 초기화
        print("\n💰 매출 현황:")
        for 빵, 판매량 in self.sales.items():
            매출 = 판매량 * self.prices[빵]  # 판매량 × 가격
            total_sales += 매출  # 총 매출에 추가
            print(f"- {빵}: {판매량}개 판매, 매출 {매출}원")
        
        print(f"\n🔹 총 매출: {total_sales}원")  # 총 매출 출력
        return total_sales

    # ✅ 매출 데이터를 파일에 저장
    def save_sales_data(self):
        data = {
            "stock": self.stock,
            "sales": self.sales,
            "total_sales": self.calculate_total_sales()
        }
        with open("sales_data.json", "w", encoding="utf-8") as file:
            json.dump(data, file, ensure_ascii=False, indent=4)
        print("\n💾 매출 데이터가 저장되었습니다. (sales_data.json)")

# ✅ 메인 함수 구현
def main():
    shop = BungeoppangShop()  # 붕어빵 가게 객체 생성

    while True:
        print("\n===== 붕어빵 가게 =====")
        print("1. 🛒 주문 모드")
        print("2. 🔧 관리자 모드")
        print("3. 💰 총 매출 확인 및 종료")
        choice = input("원하는 모드를 선택하세요 (1/2/3): ")

        if choice == "1":  # 주문 모드
            shop.check_stock()
            빵_타입 = input("주문할 붕어빵 종류를 입력하세요: ")
            빵_숫자 = int(input("몇 개를 주문하시겠어요? "))
            shop.process_order(빵_타입, 빵_숫자)

        elif choice == "2":  # 관리자 모드
            관리자_비밀번호 = "1234"
            입력_비밀번호 = input("🔑 관리자 비밀번호를 입력하세요: ")
            if 입력_비밀번호 != 관리자_비밀번호:
                print("❌ 비밀번호가 틀렸습니다!")
                continue
            
            shop.check_stock()
            빵_타입 = input("재고를 추가할 붕어빵 종류를 입력하세요: ")
            추가_재고 = int(input("추가할 개수를 입력하세요: "))
            shop.admin_mode(빵_타입, 추가_재고)

        elif choice == "3":  # 총 매출 확인 및 종료
            shop.calculate_total_sales()
            shop.save_sales_data()
            print("\n🚪 프로그램을 종료합니다. 이용해주셔서 감사합니다!")
            break  # 루프 종료

        else:
            print("❌ 올바른 번호를 입력하세요!")

```

```python
# 프로그램 실행
if __name__ == "__main__":
    main()
```
